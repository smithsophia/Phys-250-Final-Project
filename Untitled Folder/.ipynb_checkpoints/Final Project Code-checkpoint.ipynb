```python
# Import the usual numpy and matplotlib libraries for numerical operations and plotting
import numpy as np
import matplotlib.pyplot as plt
from numpy import random as rng
```

### Behind the scenes work:

```python
def randomlattice(N, M,Lp):
    return rng.choice((0, 1), (N, M), p=[1-Lp, Lp])
#print(randomlattice(5,5,.5))
```

Now I need to figure out a way how to set a matrix with a certain level of life. 

```python
def Maggy(lattice):#is this the sum of every particle in the lattice 
    N= len(lattice)
    M= len(lattice[0])
    total=0
    for i in range(N):
        for j in range(M):
            magnt=lattice[i][j]
            total+= magnt     
    return total #Returns M

#print(Maggy(harvest))
```

```python
def neighborsSum(lattice, i, j):
    N= len(lattice)
    M= len(lattice[0])
    pot=lattice[i][j]
    summy= np.sum([lattice[(i+1) % N][j], lattice[i][(j+1) % M], lattice[(i-1) % N][j], lattice[i][(j-1) % M], lattice[(i+1) % N][(j+1) % M],lattice[(i+1) % N][(j-1) % M], lattice[(i-1) % N][(j-1) % M],lattice[(i-1) % N][(j+1) % M]] )
    return [summy, lattice, pot] 
                  
#print(neighborsSum(harvest, 2,2))
```

```python
def plotty(lattice, it, name):
    fig, ax = plt.subplots()
    im = ax.imshow(lattice)
    fig.tight_layout()
    name =  name + str(it) +'.jpg'
    plt.savefig(name)
    plt.close()
```

Next up is going to be a function that flips one cell if the rules tells it to. 

Any live cell with fewer than two live neighbours dies, as if by underpopulation.
Any live cell with two or three live neighbours lives on to the next generation.
Any live cell with more than three live neighbours dies, as if by overpopulation.
Any dead cell with exactly three live neighbours becomes a live cell, as if by reproduction.

```python

def LifeOne(lattice, x, y):
    tree=lattice
    if tree[x][y] == 1 and  neighborsSum(tree, x, y)[0]<2 :
        #print('cell dies by underpopulation')
        return 0    
    elif tree[x][y] == 1 and neighborsSum(tree, x, y)[0]> 3:
        #print('cell dies by overpopulation')
        return 0 
       
    elif tree[x][y] == 0 and neighborsSum(tree, x, y)[0]>=3:
        #print('regrowth')
        return 1
    else:
        #print('nothing')
        return tree[x][y]

#print(LifeOne(harvest, 2,2))
        
```

```python
def multiLife(lattice):
    N= len(lattice)
    M= len(lattice[0])
    tree=lattice 
    newtree= np.zeros((N,M))
    for i in range(N):
        for j in range(M):
            newtree[i][j]=LifeOne(tree, i, j)#we will just build a new one not change the old one. 
            
    return(newtree)

#print(multiLife(harvest))


    
```

```python
hahaclintondix=randomlattice(50,50,.5)
def somuchlife(lattice, reps, name, saveimages= False):
    N= len(lattice)
    M= len(lattice[0])
    E= np.zeros(reps)
    lifes= np.zeros((reps, N, M))
    lifes[0]=(lattice)
    E[0]= Maggy(lattice)
    for i in range(1, reps):
        lattice = (multiLife(lattice))
        lifes[i] = lattice
        E[i]=Maggy(lattice)
        if saveimages: 
            plotty(lattice, i, name) 
    return(E, lifes)
                 
#print(somuchlife(hahaclintondix, 6, 'blank' )[0])    
```

```python
#mylat= randomlattice(5,5,.5)
def randomfit(N,M,Lp):
    life=randomlattice(N,M,Lp)
    total=np.zeros((N, M))
    randoms=rng.random_sample((N,M))
    for i in range(N):
        for j in range(M):
            total[i][j] =life[i][j]*randoms[i][j]
    return(total)

#print(randomfit(5,5,.5))
```

```python
def Population(lattice):
    N= len(lattice)
    M= len(lattice[0])
    E=0
    for i in range(N):
        for j in range(M):
            if lattice[i][j]>0:
                E+=1
    return E
```

So I have a new matrix now. I will next need to make the thing that adds up the probability and makes the new probability 

```python
#the most number of neighbors you can have is 8
#test
#x=[.5,.1, .2, .3, .4, .5, .6, .7, .8]
def relprob(arr):
    N=len(arr)
    prob= np.zeros(N)
    for i in range(len(arr)):
        prob[i]=arr[i]/sum(arr)
    return(prob)
#print(relprob(x))
```

```python
#This value chooses the outcome basesd on the prbabilities above. if cell is dead and regenerating, 0 percent chance of staying dead. 
def problife(array):
    y=rng.choice(array, p=relprob(array))
    return(y)
#print(problife(x))
```

```python
#print(plotlattice(randomfit(5,5,.6)))
def eveneighborsSum(lattice, i, j):
    N= len(lattice)
    M= len(lattice[0])
    pot=lattice[i][j]
    n=0
    p= [lattice[i][j], lattice[(i+1) % N][j], lattice[i][(j+1) % M], lattice[(i-1) % N][j], lattice[i][(j-1) % M], lattice[(i+1) % N][(j+1) % M],lattice[(i+1) % N][(j-1) % M], lattice[(i-1) % N][(j-1) % M],lattice[(i-1) % N][(j+1) % M]]
    for h in range(1,9): 
        if p[h]>0:
            n+=1
    return [p, n] 
#print(eveneighborsSum(randomfit(5,5,.5) , 2,2))
```

```python
def eveLifeOne(lattice, x, y):
    tree=lattice
    if tree[x][y] > 0  and  eveneighborsSum(tree, x, y)[1]<2 :
        #print('cell dies by underpopulation')
        return 0
    elif tree[x][y] > 0  and eveneighborsSum(tree, x, y)[1]> 3:
        #print('cell dies by overpopulation')
        probfii = problife(eveneighborsSum(lattice,x,y)[0])
        if tree[x][y]== probfii:
            return 0 #tree[x][y]
        elif tree[x][y]!= probfii:
            return 0
       
    elif tree[x][y] == 0 and eveneighborsSum(tree, x, y)[1]>=3:
        probfit=problife(eveneighborsSum(lattice,x,y)[0])
        return probfit
    else:
        #print('nothing')
        return tree[x][y]

#print(randomfit(5,5,.5))
#zz=randomfit(5,5,.5)
#print(zz)
#print(eveLifeOne(zz , 1,2))
```

```python
def eqFinder(top,  pop, name, stepsize, tol, dim=100,  plotimages= False):
    f = randomlattice(dim , dim, pop)
    avg = AboveAverage(f, top, name)
    if plotimages: 
        AvgPlot( avg, top, name, pop, dim)
    slope = tol + 1
    for i in range(int(top/stepsize)):
       
        if abs(slope) < tol:
            break 
        high = (i+1)*stepsize -1 
        low = i*stepsize
        eqspot = int((high + low) / 2) 
        slope = (avg[high] - avg[low] ) / stepsize
    if abs(slope) > tol: 
        eqachived = False
        print('Slope is greater than tolerance. Equilibrium may not have been reached in this range.')
    else: 
        eqachived = True       
    return (eqspot, avg[eqspot], eqachived)
```

```python
def full( top,  levels, stepsize, tol, dim=100 ,savegraphs= False):
    eqspots=np.zeros(len(levels))
    eq= np.zeros(len(levels))
    for i in range(len(levels)):
        print(levels[i])
        name = 'level_' + str(levels[i])
        stuff= eqFinder(top,  levels[i], name, stepsize, tol, dim)
        yn = stuff[2]
        if yn:
            eqspots[i]=  stuff[0]
            eq[i]= stuff[1]
        else:
            eqspots[i]=0
            eq[i]=0
        
            
    fig=plt.figure()
    y = eqspots
    x = levels
    ax=fig.add_subplot(111)
    ax.plot(x , y,'ro')
    ax.set_xlabel('Percent Population Starting Level')
    ax.set_ylabel('Time Equilibrium Achived')
    ax.set_title('Equilibrium Time vs. Perecent Population at Start')
    fig.tight_layout()
    if savegraphs: 
        plt.savefig('Eqspot_Vs_Population' +name)
    plt.show()
    
    fig2=plt.figure()
    h = eq
    x = levels
    ax=fig2.add_subplot(111)
    ax.plot(x , h,'ro')
    ax.set_xlabel('Percent Population Starting Level')
    ax.set_ylabel('Equilibrium Level')
    ax.set_title('Equilibrium vs. Perecent Population at Start')
    fig2.tight_layout()
    if savegraphs: 
        plt.savefig2('Eq_Vs_Population' +name)
    plt.show()
    return (eq, eqspots)

        
        
   
```

```python
def eqFinder_Fit(top,  pop, namefit, namepop, stepsize1, stepsize2, tolfit, tolpop, dim=100,  plotimages= False):
    f = randomfit(dim , dim, pop)
    content = avg_Fit(f, top, namefit)
    avg1 = content[0] #returns array of averages. 
    avg2 = content[1] #returns population
    if plotimages: 
        AvgPlot(avg1, top, namefit, pop, dim)
        AvgPlot(avg2, top, namepop, pop, dim)
    slope1 = tolfit + 1
    slope2 = tolpop + 1
    for i in range(int(top/stepsize1)):
        if abs(slope1) < tolfit:
            break 
        high1 = (i+1)*stepsize1 -1 
        low1 = i*stepsize1
        eqspot1 = int((high1 + low1) / 2) 
        slope1 = (avg1[high1] - avg1[low1] ) / stepsize1
    if abs(slope1) > tolfit: 
        eqachived1 = False
        print('Slope is greater than tolerance. Equilibrium may not have been reached in this range.')
    else: 
        eqachived1 = True       
        
    for j in range(int(top/stepsize2)):
        if abs(slope2) < tolpop:
            break 
        high2 = (i+1)*stepsize2 -1 
        low2 = i*stepsize2
        eqspot2 = int((high2 + low2) / 2) 
        slope2 = (avg2[high2] - avg2[low2] ) / stepsize2
    if abs(slope2) > tolpop: 
        eqachived2 = False
        print('Slope is greater than tolerance. Equilibrium may not have been reached in this range.')
    else: 
        eqachived2 = True 
        
    return (eqspot1, avg1[eqspot1], eqachived1 , eqspot2, avg2[eqspot2], eqachived2)
```

```python
def full_Fit( top, levels, namefit, namepop, stepsize1, stepsize2, tolfit, tolpop, dim=100 ,savegraphs= False):
    eqspots1=np.zeros(len(levels))
    eq1= np.zeros(len(levels))
    eqspots2=np.zeros(len(levels))
    eq2= np.zeros(len(levels))

    for i in range(len(levels)):
        print(levels[i])
        stuff = eqFinder_Fit(top,  levels[i], namefit, namepop, stepsize1, stepsize2, tolfit, tolpop, dim,  plotimages= False)

        yn1 = stuff[2]
        yn2 = stuff[5]
        if yn1:
            eqspots1[i] =  stuff[0]
            eq1[i] = stuff[1]
        if yn2: 
            eqspots2[i] = stuff[3]
            eq2[i] = stuff[4]
            
        elif yn1 == False:
            eqspots1[i]=0
            eq1[i]=0
            
        elif yn2 == False: 
            eqspots2[i] = 0 
            eq2[i] = 0
    print('plotting 1')       
    fig=plt.figure()
    y = eqspots1
    x = levels
    ax=fig.add_subplot(111)
    ax.plot(x , y,'ro')
    ax.set_xlabel('Percent Population Starting Level')
    ax.set_ylabel('Fitness Equilibrium Time')
    ax.set_title(' Fitness Equilibrium Time vs. Perecent Population at Start')
    fig.tight_layout()
    if savegraphs: 
        plt.savefig('Fit_Eqspot_Vs_Population')
    plt.show()
    print('plotting 2')
    fig3=plt.figure()
    y2 = eqspots2
    x = levels
    ax=fig3.add_subplot(111)
    ax.plot(x , y2,'ro')
    ax.set_xlabel('Percent Population Starting Level')
    ax.set_ylabel('Time Equilibrium Achived')
    ax.set_title('Equilibrium Time vs. Perecent Population at Start')
    fig3.tight_layout()
    if savegraphs: 
        plt.savefig('Eqspot_Vs_Population')
    plt.show()
    print('plotting 3')
    fig2=plt.figure()
    h = eq1
    x = levels
    ax=fig2.add_subplot(111)
    ax.plot(x , h,'ro')
    ax.set_xlabel('Percent Population Starting Level')
    ax.set_ylabel('Fitness Equilibrium Level')
    ax.set_title('Fitness Equilibrium  vs. Perecent Population at Start')
    fig2.tight_layout()
    if savegraphs: 
        plt.savefig('Fit_Eq_Vs_Population')
    plt.show()
    print('plotting 4')
    fig4=plt.figure()
    h2 = eq2
    x = levels
    ax=fig4.add_subplot(111)
    ax.plot(x , h2,'ro')
    ax.set_xlabel('Percent Population Starting Level')
    ax.set_ylabel('Population Equilibrium Level')
    ax.set_title('Population  Equilibrium  vs. Perecent Population at Start')
    fig4.tight_layout()
    if savegraphs: 
        plt.savefig('Pop_Eq_Vs_Population')
    plt.show()
    return (eq1, eqspots1 , eq2, eqspots2)

   
```
